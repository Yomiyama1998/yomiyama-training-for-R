```julia
c1=c2=d1=d2=1.0
u0=v0=0.7
Δt=0.05
function euler(u,v)
    u_new=(1+Δt*d1-Δt*c1*v)*u
    v_new=(1-Δt*d2+Δt*c2*u)*v
    u,v=u_new, v_new
    return u,v
end
euler(u0,v0)
```

```julia
u, v=u0, v0
uv_sq=[u0, v0]

for i in 1:500
    u,v = euler(u,v)
    uv_sq = hcat(uv_sq, [u,v] )
end
```

```julia
t_sq=[0:Δt:500*Δt]
    plot(t_sq, uv_sq')
```
