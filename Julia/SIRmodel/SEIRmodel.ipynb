```julia
using DifferentialEquations
using Plots;gr()
```

```julia
function SEIR!(du,u,p,t)
    S,E,I,R=u
    N,m,b,a,g=p
    du[1]=m*(N-S)-b*S*I
    du[2]=b*S*I-(m+a)*E
    du[3]=a*E-(m+g)*I
    du[4]=g*I-m*R
end
```

```julia
u0=[9998,1,1,0]
```

```julia
p=[10000,0.020,0.0020,0.15,0.05]
```

```julia
tspan=(0.0,200.0)
```

```julia
prob=ODEProblem(SEIR!,u0,tspan,p)
sol=solve(prob)
```

```julia
plot(sol)
```
