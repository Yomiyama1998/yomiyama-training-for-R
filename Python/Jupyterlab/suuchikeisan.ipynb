# オイラー法
$
\frac{dx}{dt}= ax 
$

$
x(t+1)=x(t)+f'(t)\times \frac{1}{n}
$

$
x=exp(at)
$

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
n = 1000 # ステップ数
dt = 0.01 # tの刻み幅
a = 1.0 # a の値

# 初期条件
x0 = 1.0
# t, x の値を格納するリスト
t_list = []
x_list = []
x_exact = [] # 解析解を格納するリスト

# t=0 では x=x0
t = 0.0
x = x0

# 値を格納する。
t_list.append(t)
x_list.append(x)
x_exact.append(x)

for i in range(n):
    f = a * x # 右辺を予め計算する。
    x = x + f * dt # Euler 法による時間発展
    t = t + dt # 時刻をdtだけ進める。

    # 計算した x, t の格納
    t_list.append(t)
    x_list.append(x)
    x_exact.append(np.exp(a*t))
```

```python
# グラフに結果を描画する。
plt.plot(t_list, x_list, label='simulation')
plt.plot(t_list, x_exact, label='analytical')
plt.xlabel('t')
plt.ylabel('x')
plt.yscale('log')
# 凡例を表示する。
plt.legend(loc='best')
```

# オイラー法
$
dx=cos(t)dt
$

$
\int dx = \int cos(t) dt
$

$
x = sin(t)+x(t=0)
$

これらをテイラー展開すると \
$
x(t+\delta t)=x(t)+\frac{x(t)}{dt}\Delta t + O(\Delta t^2)
$

$O(\delta t^2)$を打ち切り誤差として切り捨てると \
$
x(t+\delta t) \simeq x(t) + \frac{x(t)}{dt}\Delta t \\
              = x(t)+ cos(t)\Delta t
$


```python
import math
import matplotlib.pyplot as plt
import numpy as np

def f(t):
  return math.cos(t)

DELTA_T = 0.001
MAX_T = 100.0

t = 0.0 # t初期値
x = 0.0 # t=0でのx

x_hist = [x]
t_hist = [t]

# 逐次計算
while t < MAX_T:
    x += f(t)*DELTA_T
    t += DELTA_T
    x_hist.append(x)
    t_hist.append(t)

# 数値解のプロット  
plt.plot(t_hist, x_hist)
# 厳密解(sin(t))のプロット

t = np.linspace(0,MAX_T,int(1/DELTA_T))
x = np.sin(t)
plt.plot(t, x)

plt.xlim(0, MAX_T)
plt.ylim(-1.3, 1.3)

plt.show()
```

# 中心差分法
$x(t+\delta t),x(t-\delta t)$をテイラー展開すると
$$
x(t+\delta t) = x(t)+\frac{dx(t)}{dt}\delta t +\frac{1}{2} \frac{d^2x(t)}{dt}\delta t^2+O(\Delta t^3)  \\
x(t-\delta t) = x(t)-\frac{dx(t)}{dt}\delta t +\frac{1}{2} \frac{d^2x(t)}{dt}\delta t^2+O(\Delta t^3) 
$$
上式から下式を引くと
$$
x(t+\delta t) - x(t-\delta t) \simeq 2\frac{dx(t)}{dt}\Delta t \\
x(t+\delta t) \simeq x(t-\delta t)+2cos(t)\Delta t \
$$

$t+\Delta t \rightarrow t$で置き換えると
$$
x(t)\simeq x(t-2\Delta t)+2cos(t-\Delta t)\Delta t
$$
$t-2\Delta t$での$x$がわかっている時最後の式により$x$を逐次的に計算していく方法

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    return math.cos(t)

DELTA_T = 0.001
MAX_T = 100.0

t = 0.0 # t初期値
x = 0.0 # t=0でのx

x_hist = [x]
t_hist = [t]

while t < MAX_T:
    x += 2*f(t-DELTA_T)*DELTA_T
    t += 2*DELTA_T
    x_hist.append(x)
    t_hist.append(t)

# 数値解のプロット 
plt.plot(t_hist, x_hist)

# 厳密解(sin(t))のプロット
t = np.linspace(0, MAX_T, int(1/DELTA_T))
x = np.sin(t)
plt.plot(t, x)

plt.xlim(0, MAX_T)
plt.ylim(-1.3, 1.3)

plt.show()
```
